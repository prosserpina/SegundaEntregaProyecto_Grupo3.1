#### 3.1.1.1. *Distribuciones de frecuencia*

- Contamos los valores para las frecuencias absolutas 

```python
frecS = pokemon['status'].value_counts()
frecS
```

- Guardamos los datos de frecuencias absolutas en un DataFrame

```python
frecS_df = pd.DataFrame(frecS)
frecS_df
```

- Asignamos un nombre a la columna del DF creado


```python
frecS_df.rename(columns={'status':'Frecuencia Absoluta'}, inplace=True)
frecS_df
```

- Obtenemos los valores de las frecuencias absolutas acumuladas 

```python
frec_abs_val = frecS_df['Frecuencia Absoluta'].values

acumS=[]

valor_acumS=0

for i in frec_abs_val:
    valor_acumS=valor_acumS + i
    acumS.append(valor_acumS)
    
frecS_df['Frec abs acumulada'] = acumS
frecS_df
```

- Calculamos la frecuencia relativa

```python
frecS_df['Frecuencia Relativa'] = round(100 * frecS_df['Frecuencia Absoluta']/len(pokemon.status),2)
frecS_df
```

- Obtenemos el acumulado de las frecuencias relativas

```python
frec_rel_val = frecS_df['Frecuencia Relativa'].values

acum_relS = []

valor_acum_relS = 0

for i in frec_rel_val:
    valor_acum_relS= valor_acum_relS+i
    acum_relS.append(valor_acum_relS)
    
frecS_df['Frec rel acumulada'] = acum_relS
frecS_df
```

Podemos concluir que los tipo de Pokémon "Normal" representan casi el 90% del total, y el tipo "Mythical" son la especie más rara de encontrar.

### 4.2. Algoritmos de agrupamiento
El objetivo de los algoritmos de agrupamiento es encontrar una estructura o un patrón en una colección de datos no clasificados.
Intentan encontrar grupos en los datos que compartan atributos en común. 

### 4.2.1. K-means
El objetivo de este algoritmo es encontrar grupos en los datos, con el número de grupos representados por la variable K. El algoritmo funciona de manera iterativa para asignar cada punto de datos a uno de los grupos K en función de las características que se proporcionan.

```python
#K-means

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,10)

from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings("ignore")
```

```python
# Realizaremos análisis utilizando datos de las variables ataque y defensa, para tratar de establecer clusters

Xmeans=pokemon.iloc[:,[11,12]].values
Xmeans
```

```python
# Visualizamos la data contenida
plt.scatter(
   Xmeans[:, 0], Xmeans[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
```

```python
# Antes de crear el modelo debemos saber cuantos clusters crear, para eso usamos el como de jambu
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i,
        max_iter=300,
        random_state=0
    )
    km.fit(Xmeans)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
# Ya vimos que la cantidad optima de clusters es 2, asi que creamos y entrenamos el modelo

modelKM = KMeans(
    n_clusters=2,
    max_iter=300, 
    random_state=0
)
predicKM = modelKM.fit_predict(Xmeans)
```

```python
# plot the 2 clusters
plt.scatter(
    Xmeans[predicKM == 0, 0], Xmeans[predicKM == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    Xmeans[predicKM == 1, 0], Xmeans[predicKM == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)


# plot the centroids
plt.scatter(
    modelKM.cluster_centers_[:, 0], modelKM.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.title('Clusters de Pokemones')
plt.xlabel('Ataque')
plt.ylabel('Defensa')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

Los datos se encuentran desbalanceados y esto puede inducir a errores del tipo "Overfitting", lo cual implica que el modelo no estaría clasificando o generalizando apropiadamente. Hay una sobrerrepresentación de la clase "False" (Pokemon de status no Legendarios) que debe ser tratada. 

Para tratar con el Overfitting se utilizaran 2 métodos estadísticos tratando el desbalance de la data: Under-Sampling la non-legendary data & Over-Sampling la legendary data. 

#### Equilibramos la data

```python
# Son separadas las filas de Pokemon Legendarios (True) de los que no (False)
legendaryPokemon = pokemon.loc[pokemon['Legendary']== 'True']
normalPokemon = pokemon.loc[pokemon['Legendary']== 'False']

# Usaremos sólo las estadísticas base y tipos para determinar si es legendario o no
legendaryPokemon = legendaryPokemon[['type_1','type_2','total_points','hp','attack','defense','sp_attack','sp_defense','speed','Legendary']]
normalPokemon = normalPokemon[['type_1','type_2','total_points','hp','attack','defense','sp_attack','sp_defense','speed','Legendary']]

# Tomaremos muestras aleatorias de Pokemon no legendarios del conjunto de datos para equilibrar nuestro dataset

sampledNormalPokemon = normalPokemon.sample(100)
sampledNormalPokemon
```

```python
# Se vuelve a unir la data separada anteriormente
x = pd.concat([legendaryPokemon, sampledNormalPokemon])

# Tomaremos la última columna Legendary como target y la eliminaremos de la data de entrenamiento
y = x['Legendary']
x = x.drop('Legendary', 1)
```

```python
# Existiendo distintos tipos en formato texto que generarian conflicto con los metodos de clasificacion estos seran transformados a valores numericos mediante la expresion get_dummies()
x = pd.get_dummies(x)
x
```

#### División de la Data

- Importamos los paquetes/librerias necesarios: 

```python
from sklearn.model_selection import train_test_split
```

- Usamos train_test_split para crear sets de entrenamiento y testeo

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 47, test_size = 0.30)
```

En pasos anteriores se vio que la data se encuentra desbalanceada por lo que hay que realizar un ajuste de la data mediante el algoritmo SMOTE para realizar oversampling en los datos de los que se tiene menor cantidad

#### Algoritmo SMOTE

Aumentaremos la cantidad de Pokemon Legendarios en nuestros datos de entrenamiento, creando ejemplos sintéticos usando algoritmo SMOTE, a un ratio de 1.0. 

- Importamos los paquetes/librerias necesarios:



```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=15,sampling_strategy= 1.0)

oversample = SMOTE()
x, y = oversample.fit_resample(x, y)
```

#### Seteo de la nueva data de entrenamiento

```python
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

- Importamos los paquetes/librerias necesarios:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

### 4.3.1. Random Forest

```python
rf = RandomForestClassifier(max_features=5, n_estimators=500)
rf.fit(X_train_res, y_train_res)
```

### 4.3.2. Support Vector Machines

```python
svm = SVC(C = 1000, kernel = 'rbf', probability=True, random_state=47)
svm.fit(X_train_res, y_train_res)
```

### 4.3.3. Regresión Logistica

```python
log = LogisticRegression(solver='lbfgs', max_iter=4000, random_state=47)
log.fit(X_train_res, y_train_res)
```

### 4.3.4. KNN Vecinos Cercanos

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_res, y_train_res)
```

### 4.3.5. Arbol de Decisión

```python
tree = DecisionTreeClassifier(max_depth=5)
tree.fit(X_train_res, y_train_res)
```

Predicciones

```python
r_probs = [0 for _ in range(len(y_test))]
rf_probs = rf.predict_proba(X_test)
svm_probs = svm.predict_proba(X_test)
log_probs = log.predict_proba(X_test)
knn_probs = knn.predict_proba(X_test)
tree_probs = tree.predict_proba(X_test)
```

Mantenemos los resultados positivos

```python
rf_probs = rf_probs[:, 1]
svm_probs = svm_probs[:, 1]
log_probs = log_probs[:, 1]
knn_probs = knn_probs[:, 1]
tree_probs = tree_probs[:, 1]
```

### 4.4. Curva ROC y área bajo la curva

- Importamos los paquetes/librerias necesarios:

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

#### 4.4.1. Area bajo la curva AUROC

```python
r_auc = roc_auc_score(y_test, r_probs)
rf_auc = roc_auc_score(y_test, rf_probs)
svm_auc = roc_auc_score(y_test, svm_probs)
log_auc = roc_auc_score(y_test, log_probs)
knn_auc = roc_auc_score(y_test, knn_probs)
tree_auc = roc_auc_score(y_test, tree_probs)
```

#### 4.4.2. Valores de AUROC

```python
print('Random (chance) Prediction: AUROC = %.3f' % (r_auc))
print('Random Forest: AUROC = %.3f' % (rf_auc))
print('SVM (Support Vector Machines): AUROC = %.3f' % (svm_auc))
print('Regresión Logistica: AUROC = %.3f' % (log_auc))
print('KNN - Vecinos Cercanos: AUROC = %.3f' % (knn_auc))
print('Arbol de decisión: AUROC = %.3f' % (tree_auc))
```

- Parche para convertir los valores True y false en la columna y_test de tipo object a valores numericos int

```python
result = y_test.replace(to_replace = ['True', 'False'], value = [1, 0])
result = result.array
```

#### 4.4.3. Calculo de la curva ROC

```python
r_fpr, r_tpr, _ = roc_curve(result, r_probs)
rf_fpr, rf_tpr, _ = roc_curve(result, rf_probs)
svm_fpr, svm_tpr, _ = roc_curve(result, svm_probs)
log_fpr, log_tpr, _ = roc_curve(result, log_probs)
knn_fpr, knn_tpr, _ = roc_curve(result, knn_probs)
tree_fpr, tree_tpr, _ = roc_curve(result, tree_probs)
```

#### 4.4.4. Plot de de la curva ROC

- Importamos los paquetes/librerias necesarios:

```python
import matplotlib.pyplot as plt
```

- Configuración del plot y su display

```python
plt.plot(r_fpr, r_tpr, linestyle='--', label='Random prediction (AUROC = %0.3f)' % r_auc)
plt.plot(rf_fpr, rf_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % rf_auc)
plt.plot(svm_fpr, svm_tpr, marker='.', label='SVM (AUROC = %0.3f)' % svm_auc)
plt.plot(log_fpr, log_tpr, marker='.', label='Regresión logistica (AUROC = %0.3f)' % log_auc)
plt.plot(knn_fpr, knn_tpr, marker='.', label='KNN Vecinos cercanos (AUROC = %0.3f)' % knn_auc)
plt.plot(tree_fpr, tree_tpr, marker='.', label='Arbol de decisión (AUROC = %0.3f)' % tree_auc)

# Titulo
plt.title('ROC Plot')
# Etiqueta de ejes
plt.xlabel('Ratio de falsos positivos')
plt.ylabel('Ratio de verdaderos positivos')
# Mostrar leyenda
plt.legend() # 
# Mostrar plot
plt.show()
```
